    <p>Este tipo de problema tambien es llamado como <b>Grafo de visibilidad</b></p>
    <p>Un grafo de visibilidad está formado por:<p>
    <ul>
    <li>N nodos (inicio, fin y vértices de obstáculos)</li>
    <li>Conexiones entre los nodos (caminos conectores de nodos visibles entre si)</li>
    </ul>
    De  este  modo,  dos  nodos  están  conectados  si  y  solo  si  son “visibles”,  es  decir,  se  puede alcanzar el segundo nodo desde el primero (o viceversa) al seguir la línea recta que los une, sin interceptar algún obstáculo del entorno. También se consideran visibles si  el  segmento  que  une los dos nodos yace sobre una arista del polígono que modela a un obstáculo.
</p>
